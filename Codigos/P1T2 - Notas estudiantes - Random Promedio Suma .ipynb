## Datos del Ejercicio

Crear un cuaderno de Jupiter Notebook en Python 3 donde se genere de forma aleatoria las notas (entre 0.0 y 5.0) de 40 estudiantes de una asignatura, con esa informacion calcular:
2. Cuantos estudiantes perdieron (menor o igual a 2.9).
3. Cuantos Pasaron (3 o más).
4. Cual fue la nota promedio.
5. El programa no debe sobrepasar las 10 líneas de Código.




## Solución del ejercicio

### Solución Tradicional

```python
import random #Libreria para numeros random
cont = 1
notas = [ random.randrange(5.0) for i in range(40) ]
total = aprub = deprub = 0
for nota in notas:
  print("Estudiante ("+str(cont)+") nota: "+str(nota))
  if nota <= 2.9:
    print("Estudiante ("+str(cont)+") REPROBO la Materia \n")
    deprub = deprub + 1
  else:
    print("Estudiante ("+str(cont)+") APROBO la Materia \n")
    aprub = aprub + 1
  cont = cont + 1
total = sum(notas)
print("Promedio de Notas: "+str(total/40)
      +"\nNumero de Aprobados : "+str(aprub)
      +"\nNumero de Reprobados: "+str(deprub))
# No era necesario mostrar La nota individual de cada estudiante :v
```
