### Solucion

#### Punto Uno

```python
# P1T3_A
bruto = input('Cual es el valor bruto del producto: ')
print(f_aplica_iva(bruto, iva))
"""
La variable "bruto" debe especificarse que sea un número.

El valor del IVA es de 0.19, no de 19.

La sobreescritura del parámetro de la función no se da de esta manera,
hay que enviarla desde el lugar de invocación.

El orden de la estructura está mal.
Debido a que la ejecución es secuencial, en la línea tres preguntara
por la variable "f_aplica_iva" que corresponde a una función en el código,
sin embargo, al momento de leer la línea 3, aun no existe la misma, por lo
que arrojara el error de "variable no definida con antelación" y como
no existe hasta este momento, no sabe que llamar en el programa.
"""
def f_aplica_iva(bruto, iva = 19):
    total_iva = bruto * iva   
    return total_iva
```

El arreglo sería el siguiente:

```python
bruto = int(input('Cual es el valor bruto del producto: '))

def f_aplica_iva(bruto, iva):
    total_iva = bruto * iva   
    return total_iva

print(f_aplica_iva(bruto, 0.19))
```

#### Punto Dos

```python
# P1T3_B
u = (1, 2, 3)
v = (4, 5, 6)

def multiplicar_vector(u, v):
    for i in u:
        u[i+1] *= v[i+1]
    return sum(u)

"""
La multiplicación de vectores, para el programa serian tuplas de números, se deben
recorrer cada una de estas tuplas por separado en for anidados y ya una vez hecho 
esto, se puede devolver el valor en otra lista.
Cada uno de los elementos de "u" se multiplicará con cada uno de los elementos
de "v", aunque se especifique esta secuencia de multiplicación, la forma en que 
se escribe no está soportada. Prefiero sacar una lista externa donde guardar el 
resultado de la multiplicación de vectores.
Sum() es una función que permite devolver la suma de los componentes que tiene 
una lista/tupla, no está claro si se quiere ese total o los resultados por separado,
pero, asumiendo que sea así se esclarecerá como la suma de la nueva lista "w".
Un último detalle es que "u" y "v" son tuplas, esto quiere decir que su contenido
es inmutable, no se puede cambiar, mientras que "w" es una lista a la que podemos 
almacenarle nuevos elementos, es decir, podemos transformar y cuya propiedad es de
provecho en el programa.
"""

print(multiplicar_vector(u, v))
```

El arreglo seria el siguiente:

```python
u = (1, 2, 3)
v = (4, 5, 6)
w = []

def multiplicar_vector(u, v):
    for i in u:
      for j in v:
        w.append(i*j)
    return sum(w)

print(multiplicar_vector(u, v))
```
