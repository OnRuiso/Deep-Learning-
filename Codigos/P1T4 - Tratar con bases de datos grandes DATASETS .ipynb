## Crear tabla de datos

```python
# Se crea una tabla con filas=data   columnas=nombresCaracteristicos
boston = pd.DataFrame(boston_dataset.data,columns=boston_dataset.feature_names)
boston.head(10)
```

```python
# agregar en la tabla los datos del target
boston["MEDV"]=boston_dataset.target #agrega una columa al final llamada MEDV con los datos de tarjeta
boston.head(10)
```

## Graficos de datos

```python
sns.set(rc={"figure.figsize":(11.7,8.27)})#tamaño del grafico
sns.distplot(boston["MEDV"]) #Se agrega los datos a la grafica
plt.show() #Se muestra el grafico
```

## Matrix de Correlacion de caracteristicas

```python
correlation_matrix = boston.corr().round(2)
##grafico de correlacion de pandas
## mientras el valor cea mas cercano a 1 mejor correlacion tendra
## mientras mas alejado de 1 peor correlacion tendra
sns.heatmap(data=correlation_matrix,annot=True)
#graficara utilizando sns
## annot = True ; imprimira los valores dentro del cuadro
```

El coeficiente de correlacion oscila entre -1 y 1.
Si el valor esta cerca de 1 , significa que hay una fuerte correlacion positiva entre las dos variables-
Si el valor esta cerca de -1 , las variables tienen una fuerte correlacion negativa.
