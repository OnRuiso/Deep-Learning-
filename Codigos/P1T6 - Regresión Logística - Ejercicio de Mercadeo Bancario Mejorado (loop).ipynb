```python
X = df.drop(columns='si_no')
Y = df.si_no

baloto, porcentaje = 0, 0.1

while True:
  baloto = baloto + 1;
  print("random_state: "+str(baloto)+" test_size: "+str(porcentaje))
  X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = porcentaje, random_state=baloto)
  regresion_logistica = LogisticRegression()
  regresion_logistica.fit(X,Y)
  prediccion = regresion_logistica.predict(X_test)
  exactitud = ( metrics.accuracy_score(Y_test, prediccion) )*100

  if baloto == 100:
    porcentaje = porcentaje + 0.1
    baloto = 0
  
  elif porcentaje == 0.99:
    print("No se hayo un valor confiable")
    break

  elif exactitud >= 90:
    print("\n\n\n-------------------------------------------------------------------")
    print("CONFIABILIDAD= "+str(exactitud)+
          " \n \n Hayado con los siguientes Parametros: \n Test_size:"
          +str(porcentaje)+" \n Random_state: "+str(baloto))
    print("-------------------------------------------------------------------\n\n\n")
    break





```
